```python
import os
import odooly
import pandas as pd
from pandas_ods_reader import read_ods
from tqdm.notebook import tqdm
```

```python
odooly.Client._config_file = os.path.expanduser('~/.config/odooly.ini')
odoo = odooly.Client.from_config('odoo14')
```

```python
file = read_ods("unit_packing.ods")
file = pd.DataFrame(file)
```

```python
for idx,line in tqdm(file.iterrows(), total=file.shape[0]):
    item = odoo.env['product.product'].with_context(test_active=False).search([('default_code', '=',int(line['code']))]).name
    if not item:
        item = odoo.env['product.template'].with_context(test_active=False).search([('default_code', '=',int(line['code']))]).name
    print(item)
```

```python
for idx,line in tqdm(file.iterrows(), total=file.shape[0]):
    item = odoo.env['product.product'].search([('default_code', '=',int(line['code']))])
    if not item:
        item = odoo.env['product.template'].search([('default_code', '=',int(line['code']))])
    if item:
        item.write({'unit_packing':float(line['unit_packing'])})
        print(item.unit_packing)
```
