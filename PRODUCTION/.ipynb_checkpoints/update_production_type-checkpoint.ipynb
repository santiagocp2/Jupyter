```python
import os
import odooly
import pandas as pd
from pandas_ods_reader import read_ods
from tqdm.notebook import tqdm
```

```python
odooly.Client._config_file = os.path.expanduser('~/.config/odooly.ini')
odoo = odooly.Client.from_config('odoo14')
```

```python
products=odoo.env['product.product'].search([('uom_id','=',1),('type','!=','product')])
for product in products:
        product.write({'production_type':''})
        print(product.name)
```

```python
products=odoo.env['product.product'].search([('uom_id','=',3),('type','=','product'),'name like Cinta'])
print(len(products))
for product in products:
        product.write({'production_type':'product_process'})
        print(product.name)
```

```python
products=odoo.env['product.product'].search([('uom_id','=',3),('type','=','product'),'name like Bobina'])
print(len(products))
for product in products:
        product.write({'production_type':'raw_material'})
        print(product.name)
```

```python
products=odoo.env['product.product'].search([('uom_id','=',1),('type','=','product')])
print(len(products))
for product in products:
        product.write({'production_type':'product_finish'})
        print(product.name)
```
